## Desafio 1

```python
import pandas as pd
```

```python
df = pd.read_csv('raw.githubusercontent.com_alura-cursos_bibliotecas_visualizacao_main_Dados_imigrantes_canada.csv')
```

```python
df.set_index('Pa√≠s', inplace = True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
brasil = df.loc['Brasil', anos]
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
argentina = df.loc['Argentina', anos]
argentina_dict = {'ano': argentina.index.tolist(), 'imigrantes': argentina.values.tolist()}
dados_argentina = pd.DataFrame(argentina_dict)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.plot(dados_argentina['ano'], dados_brasil['imigrantes'])
plt.xticks(['1980', '1985', '1990','1995', '2000', '2005', '2010'])
```

```python
df_comparacao = df.loc[['Brasil', 'Argentina'], anos]
```

```python
df_comparacao = df_comparacao.T
df_comparacao.head()
```

```python
plt.figure(figsize = (8, 4))
plt.plot(df_comparacao['Brasil'], label = 'Brasil')
plt.plot(df_comparacao['Argentina'], label = 'Argentina')
plt.title('Imigra√ß√£o do Brasil e Argentina para o Canad√°')
plt.xlabel('Ano')
plt.ylabel('N√∫mero de imigrantes')
plt.xticks(['1980', '1985', '1990','1995', '2000', '2005', '2010'])
plt.legend
plt.show
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

```

```python
df = pd.DataFrame(vendas_2022, index = lojas)
```
